```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import pygsheets
from sbn_yt_pu_crawler import Video, Channel, Search
```

```python
# 티어링 시트 가져오기 : 시트ID로
gc = pygsheets.authorize(service_account_file='creds.json')
sheet_id = '14d-7E9FVLLp52-FtO674R4NG8moB_OeZt6gbvYNj_9w'
wks = gc.open_by_key(sheet_id)
sh = wks.worksheet_by_title('크롤링 필요 (2)') # 크롤링 필요한 시트
tiering = sh.get_as_df(has_header=True , index_column=None , start='A1', end='H100000' , numerize=True , empty_value=np.nan)
tiering
```

```python
# data = Channel.Videos(channel_id='UC1-lwASmWqqU1V_WLNIRALw').extract()
# df = pd.DataFrame(data)
# df
```

```python
channel_id_list = tiering[' 채널ID'].unique().tolist()
result_df = pd.DataFrame()

for i in tqdm(range(len(channel_id_list))):
    try:
        data = Channel.Videos(channel_id=channel_id_list[i]).extract() # limit=100, sort='newest' or 'oldest' or 'popular', target='3달 전' or ['3달 전', '4달 전']
        df = pd.DataFrame(data)
        df['channel_id'] = channel_id_list[i]
        result_df = pd.concat([result_df, df])
    except Exception as e:
        print("크롤링 오류:", e)

result_df
```

```python
result_df['published'].unique()
```

```python
# 2023.01.01 ~ 2023.08.31 사이 영상만 추출
filtered_df = result_df[~result_df['published'].str.contains('년 전')]
filtered_df = filtered_df[~filtered_df['published'].str.contains('시간 전')]
filtered_df = filtered_df[~filtered_df['published'].str.contains('분 전')]
filtered_df = filtered_df[~filtered_df['published'].str.contains('일 전')]
filtered_df = filtered_df[~filtered_df['published'].str.contains('1개월 전')] # 9월 제외
filtered_df
```

```python
filtered_df = filtered_df[['channel_id','video_id','title','published','views']]
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```
